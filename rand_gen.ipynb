```python
import pandas as pd
import numpy as np
import json
import os
```

```python
def generate_random_weights(base_weights, variation=0.1):
    """
    Создает случайные веса на основе базовых весов, добавляя случайную вариацию.
    """
    random_variation = np.random.uniform(-variation, variation, len(base_weights))
    new_weights = base_weights * (1 + random_variation)
    new_weights = new_weights / new_weights.sum()  # Нормализуем, чтобы сумма весов была равна 1
    return new_weights.tolist()

def generate_random_submissions(train, test, num_submissions=5, use_weights=True, output_dir="submissions", variation=0.1):
    # Вычисляем начальные веса для классов на основе распределения в train
    base_weight = train['class_id'].value_counts(normalize=True)
    classes = base_weight.index.tolist()  # классы

    # Создаем директорию для сохранения сабмитов и параметров, если не существует
    os.makedirs(output_dir, exist_ok=True)

    # Параметры для каждого сабмита
    submission_params = {}

    for i in range(num_submissions):
        # Генерируем случайные веса с вариацией
        if use_weights:
            weights = generate_random_weights(base_weight.values, variation)
        else:
            weights = None  # равномерное распределение, если use_weights=False

        # Генерируем предсказания
        if use_weights:
            predicted_classes = np.random.choice(classes, len(test), p=weights)
        else:
            predicted_classes = np.random.choice(classes, len(test))

        # Обновляем колонку predicted_class в test
        test['predicted_class'] = predicted_classes

        # Сохраняем сабмит в CSV
        submission_filename = f"submission_{i}_weights_{use_weights}.csv"
        submission_path = os.path.join(output_dir, submission_filename)
        test.to_csv(submission_path, index=False)

        # Сохраняем параметры для этого сабмита
        submission_params[f"submission_{i}"] = {
            "use_weights": use_weights,
            "weights": weights if use_weights else "uniform",
            "filename": submission_filename
        }

    # Сохраняем параметры всех сабмитов в JSON-файл
    params_path = os.path.join(output_dir, "submission_params.json")
    with open(params_path, 'w') as f:
        json.dump(submission_params, f, indent=4)

    print(f"Generated {num_submissions} submissions in '{output_dir}' directory.")
```

```python
# Генерация сабмитов с параметрами
generate_random_submissions(train, test, num_submissions=10, use_weights=True, variation=0.9)
```
