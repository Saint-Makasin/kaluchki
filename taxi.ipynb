```python
def extract_datetime_features(df, datetime_col):
    # Преобразование столбца в формат datetime, если это еще не сделано
    df[datetime_col] = pd.to_datetime(df[datetime_col], errors='coerce')

    # Извлечение базовых признаков
    df['year'] = df[datetime_col].dt.year
    df['month'] = df[datetime_col].dt.month
    df['day'] = df[datetime_col].dt.day
    df['day_of_week'] = df[datetime_col].dt.dayofweek
    df['hour'] = df[datetime_col].dt.hour

    # Бинарные признаки
    df['is_weekend'] = df['day_of_week'].isin([5, 6]).astype(int)  # Суббота и воскресенье
    # df['is_holiday'] = df[datetime_col].dt.strftime('%m-%d').isin(['01-01', '07-04', '12-25']).astype(int)  # Пример праздников

    # Циклические признаки
    df['hour_sin'] = np.sin(2 * np.pi * df['hour'] / 24)
    df['hour_cos'] = np.cos(2 * np.pi * df['hour'] / 24)
    df['month_sin'] = np.sin(2 * np.pi * df['month'] / 12)
    df['month_cos'] = np.cos(2 * np.pi * df['month'] / 12)

    return df
```

```python
train1 = extract_datetime_features(train1, 'pickup_datetime')
```

```python
from geopy.distance import geodesic
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd

def process_geospatial_features(df, pickup_lat_col, pickup_lon_col, dropoff_lat_col, dropoff_lon_col, center_coords=(40.7143, -74.006), n_clusters=2):
    '''# Расчет расстояния от точки начала поездки до центра NY
    df['pickup_distance_to_center'] = df.apply(
        lambda row: geodesic((row[pickup_lat_col], row[pickup_lon_col]), center_coords).km, axis=1
    )
    
    # Расчет расстояния от точки завершения поездки до центра NY
    df['dropoff_distance_to_center'] = df.apply(
        lambda row: geodesic((row[dropoff_lat_col], row[dropoff_lon_col]), center_coords).km, axis=1
    )'''

    # Кластеризация K-средних для начальных точек поездки
    kmeans_pickup = KMeans(n_clusters=n_clusters)
    df['pickup_cluster'] = kmeans_pickup.fit_predict(df[[pickup_lat_col, pickup_lon_col]])

    # Кластеризация K-средних для конечных точек поездки
    kmeans_dropoff = KMeans(n_clusters=n_clusters)
    df['dropoff_cluster'] = kmeans_dropoff.fit_predict(df[[dropoff_lat_col, dropoff_lon_col]])

    # Признаки синус и косинус для направления начальных точек поездки
    df['pickup_lat_sin'] = np.sin(np.radians(df[pickup_lat_col]))
    df['pickup_lat_cos'] = np.cos(np.radians(df[pickup_lat_col]))
    df['pickup_lon_sin'] = np.sin(np.radians(df[pickup_lon_col]))
    df['pickup_lon_cos'] = np.cos(np.radians(df[pickup_lon_col]))

    # Признаки синус и косинус для направления конечных точек поездки
    df['dropoff_lat_sin'] = np.sin(np.radians(df[dropoff_lat_col]))
    df['dropoff_lat_cos'] = np.cos(np.radians(df[dropoff_lat_col]))
    df['dropoff_lon_sin'] = np.sin(np.radians(df[dropoff_lon_col]))
    df['dropoff_lon_cos'] = np.cos(np.radians(df[dropoff_lon_col]))

    return df
```

```python
train1 = process_geospatial_features(train1, 'pickup_latitude', 'pickup_longitude', 'dropoff_latitude', 'dropoff_longitude')
```

```python
features = train1.drop(columns=['pickup_datetime', 'key', 'fare_amount']).columns.tolist()
target = 'fare_amount'

# Заполнение пропусков
imputer = SimpleImputer(strategy='mean')
train1[features] = imputer.fit_transform(train1[features])
X_train, X_test, y_train, y_test = train_test_split(train1[features], train1[target], test_size=0.2, random_state=42)
```

```python
# Определение моделей для ансамбля
model_cb = CatBoostRegressor(iterations=500, verbose=100)  # verbose=100 для CatBoost
model_rf = RandomForestRegressor(n_estimators=40, random_state=42, n_jobs=-1)
model_lr = LinearRegression()
model_dt = DecisionTreeRegressor(max_depth=10, random_state=42)
model_knn = KNeighborsRegressor(n_neighbors=10, n_jobs=-1)
#model_xgb = XGBRegressor(learning_rate=2e-5, n_estimators=100, random_state=42, verbosity=2, n_jobs=-1)
model_lgbm = LGBMRegressor(learning_rate=3e-4, n_estimators=100, random_state=42, verbose=-1, n_jobs=-1)

# Создаем VotingRegressor с "средним" голосованием
ensemble_model = VotingRegressor(estimators=[
    ('catboost', model_cb),
    ('randomforest', model_rf),
    ('linear', model_lr),
    ('decisiontree', model_dt),
    ('knn', model_knn),
#    ('xgboost', model_xgb),
    ('lightgbm', model_lgbm)
])

# Обучение ансамбля
ensemble_model.fit(X_train, y_train)
```

```python
# Предсказание и оценка ансамбля
y_pred_ensemble = ensemble_model.predict(X_test)
rmse_ensemble = mean_squared_error(y_test, y_pred_ensemble, squared=False)
print("RMSE Voting Ensemble (CatBoost, RandomForest, etc.):", rmse_ensemble)
```
